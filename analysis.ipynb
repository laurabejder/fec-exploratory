Looking at the PAC summaries

```python
naral_sum = pd.read_csv('data/naral_pac_summary_2010_2022.csv')
emily_sum = pd.read_csv('data/emilys_list_pac_summary_2010_2022.csv')
parenthood_sum = pd.read_csv('data/planned_parenthood_pac_summary_2010_2022.csv')
```

```python
total_pac = naral_sum.append(emily_sum, ignore_index=True)
total_pac = total_pac.append(parenthood_sum, ignore_index=True)
total_pac.groupby('file_year').sum().sort_values('ttl_receipts', ascending=False)
```

```python
total_national_pac = total_pac.loc[total_pac['cmte_nm'].isin(["EMILY'S LIST",'WOMEN VOTE!','PLANNED PARENTHOOD VOTES','NARAL PRO-CHOICE AMERICA','PLANNED PARENTHOOD ACTION FUND INC. PAC','PLANNED PARENTHOOD ACTION FUND INC PAC, DBA PLANNED PARENTHOOD FEDERAL PAC','PLANNED PARENTHOOD ACTION FUND INC. PAC, DBA PLANNED PARENTHOOD FEDERAL PAC'])]
```

```python
total_national_pac.groupby('file_year').sum().sort_values('ttl_receipts', ascending=False)
```
